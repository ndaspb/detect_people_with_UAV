```python
for i in range(len(df)):
    x1 = int(df.loc[i]['x']*x-df.loc[0]['w']*x/2)
    x2 = int(df.loc[i]['x']*x+df.loc[0]['w']*x/2)
    y1 = int(df.loc[i]['y']*y-df.loc[0]['h']*y/2)
    y2 = int(df.loc[i]['y']*y+df.loc[0]['h']*y/2)
    cv2.rectangle(image, (x1, y1), (x2, y2), (0,0,255), 10)
# print(x1,x2,y1,y2)
```
